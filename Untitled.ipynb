```python
msg = messages.GetLast()
count=0
while msg:
    # Get email date 
    date = msg.ReceivedTime.strftime("%d-%m-%Y")
    # Get Subject Line of email
    #print(str(date).split("/")[0])
    if FormatedDate == date:
        sjl = msg.Subject
        # Set the critera for whats wanted                       
        #if d == date and msg.Subject.startswith("xx") or msg.Subject.startswith("yy"):
        print(str(count) +" Date : ", date + " Subject: " + sjl + " sender " + msg.SenderName)
    else:
        break
    count+=1
    msg = messages.GetPrevious()
```
